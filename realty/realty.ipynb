## Предобработка данных

Начнем обработку с потолков:

```python
# Стандартные высоты потолков начинаются с 1.64 м и заканчиваются 3.22, остальные значения не являются типовыми для застройщиков, но имеют место быть в частных случаях. 
#display(dr[dr['ceiling_height'] >5]) #по выборке видно, что с 10 м значения высоты потолков становятся уникальными.
# Поэтому отдельно изучим выбросы выше 10 и ниже 2.Возможно высокие здания - частные дома
# высота потолков в 10 метров, может относиться к частным домам, ошибочно указанной высоте дома или просто быть неверным значением. Можно их откинуть выборкой. Выше 24 – опечатка по запятой.
display(dr[dr['ceiling_height'] < 2]) # выборка кавртир с потолками ниже 2 м, выбросы или неверные значения, если потолок ниже 1.6м.
display(dr[dr['ceiling_height'] >10]) # выборка кавртир с потолками больше 10 м, выбросы или неверные значения
display(dr[dr['airports_nearest'] == 34963]) # выброс с 5 балконами и 100 метрами потолка
display(dr[dr['airports_nearest'] == 14814]) # 14 м – это дублированная площадь кухни, неизвестна высота потолка по дому, далее заменем медианой.
display(dr[dr['airports_nearest'] == 36353]) #(10,3 м) расчет среднего потолка для дома 2,8м
```

```python
# замещение неверных значений
# меняем на приемлимое для частного дома значение. С 24 меняем на точку 2.х, с 10-14 хз чд. 100 похоже на выборос с 5 балконами. 14 похоже на дубликат количества этажей, найти дубли по расстоянию и 10,3 это кухня, так высота потолка 2,8, меньше 2 частные дома – выброс.
# создаём отдельное значение с медианой дома, чтобы она не перезаписывалась при подстановке. 
h36353_mean=dr[(dr['airports_nearest'] == 36353) & dr['ceiling_height'] < 4]['ceiling_height'].mean()
dr['ceiling_height'] =  dr['ceiling_height'].replace(10.30,h36353_mean)
#по той же логике заполним пропуски этого дома
# заполняем пропуски медианой в столбце ceiling_height
dr.loc[19351,'ceiling_height'] = h36353_mean
dr.loc[20142,'ceiling_height'] = h36353_mean

#display(dr.loc[dr.loc[:,'airports_nearest'] == 36353 ]['ceiling_height']) #проверка

# проверка среза с 0 комнат, возможно это студии, либо однокомнатные квартиры, если есть кухня в отдельной площади.
#display(dr[dr['rooms']==0])
#display(dr[dr['rooms']==0].count()) # значения столбца is_studio и kitchen_area подтверждают, что 0 комнат в квартирах студиях.

# заменяем 0 комнат на 1, так как по информации из датафрейма видно, что это студии.
dr['rooms'] = dr['rooms'].replace(0,1)
#заменяем слишком маленькую цену в городе на более вероятную.
dr['last_price'] = dr['last_price'].replace(12190,12190000)

#приводим высоту потолка >20м к нормальному значению.
dr.loc[dr['ceiling_height'] >=20, 'ceiling_height'] = dr.loc[dr['ceiling_height'] >=20, 'ceiling_height'] / 10
display(dr[dr['ceiling_height'] >10]) # в дальнейшем заменить 14 м на среднюю, вместе с Nan Не делю на 10, так как планирую все значения ниже 1,70 откинуть как выбросы, а эту строчку сохраним в ДР.

#найдем границы для отнесения потолков по группам
dr['ceiling_height'].describe()

adv_mean = 2.6979 #значения получены методом describe

adv_std = 0.237247

#разносим высоту потолков по группам
adv_max_ceiling = adv_std*3+adv_mean
adv_min_ceiling = (adv_std*(-3)+adv_mean)

print(f'Высокие потолки начинаются с {adv_max_ceiling}')
print(f'Средние потолки начинаются с {adv_min_ceiling}')

#dr.plot(y='ceiling_height',kind='bar')
dr['ceiling_height'].hist(bins=80,range=(1.9,3.5)) # выборка с 2.4 по 3.5м.
display(dr['ceiling_height'].describe())
#медиана для потолков
median_ceiling_height = dr.loc[dr['ceiling_height']!=14,'ceiling_height'].median()


display(median_ceiling_height)
dr['ceiling_height'].fillna(median_ceiling_height,inplace=True)
dr['ceiling_height'] = dr['ceiling_height'].replace(14,median_ceiling_height)
#проверка замен
#dr['ceiling_height'].isna().sum() #проверка замены всех значений в ceiling_height
#display(dr.loc[355]) #проверка строки

#создаем категории по столбцу ceiling_height, для упрощения ориентирования по высоте потолков
def group_ceiling(height):
    if height >=adv_max_ceiling:
        return 'высокий'
    if adv_max_ceiling>height >=adv_min_ceiling:
        return 'средний'
    if height < adv_min_ceiling:
        return 'низкий'
dr['ceiling_height_group'] = dr['ceiling_height'].apply(group_ceiling)

#создаем категории по столбцу floor, для упрощения ориентирования по этажу
def group_floor(floor,floors): 
    if floor ==floors:
        return 'последний'
    if floors>floor >1:
        return 'другой'
    if floor == 1:
        return 'первый'
# создаем столбец с типом этажа    
dr['floor_group'] = dr.apply(lambda x: group_floor(x['floor'],x['floors_total']),axis=1)

#заполняем пустые значения медианами
#заполняем по группам по количеству комнат столбец жилой площади

for rooms in dr['rooms'].unique():
    median = dr.loc[dr['rooms'] == rooms, 'living_area'].median()
    print(rooms,median)
    dr.loc[(dr['living_area'].isna()) & (dr['rooms']==rooms), 'living_area']= median
print('Пропуски после:', dr['living_area'].isna().sum())

#заполняем по группам по количеству комнат столбец кухонной площади.
for rooms in dr['rooms'].unique():
    median = dr.loc[dr['rooms'] == rooms, 'kitchen_area'].median()
    print(rooms,median)
    dr.loc[(dr['kitchen_area'].isna()) & (dr['rooms']==rooms), 'kitchen_area']= median
print('Пропуски после:', dr['kitchen_area'].isna().sum())

#замена общей медианой пустых значений в parks_around3000 и ponds_around3000.
dr.loc[dr['parks_around3000'].isna(),'parks_around3000'] = dr['parks_around3000'].median()     
dr.loc[dr['ponds_around3000'].isna(),'ponds_around3000'] = dr['ponds_around3000'].median() 
#display(dr['parks_around3000'].isna().sum()) #проверка
#display(dr['ponds_around3000'].isna().sum()) #проверка


#print(dr.head()) #проверка
```

```python
# заполнение пропусков
#пустые значения в is_apartment скорее остались от ручного значения, где вместо False предполчи поставить пропуск.
dr['is_apartment'] = dr['is_apartment'].fillna(False)
#пустые значения в балконах наверняка возникли по тому же принципу, что и в аппартаментах.
dr['balcony'] = dr['balcony'].fillna(0)

# display(dr.info()) проверка отсутствия пропусков

# цикл для приведения к нижнему регистру столбцов из списка list_one
for lower in dr: 
    if 'locality_name' in lower:
        dr[lower] = dr[lower].str.lower()
display(dr['locality_name'].value_counts().sort_values()) # проверка 

#приведение locality_name к типу данных object
dr['locality_name'] = dr['locality_name'].astype('str')

from pymystem3 import Mystem
m = Mystem() 
from collections import Counter

# объединим уникальные значения по столбцу locality_name в строку
lemma= " ".join(dr['locality_name'].unique()) 

# лемматизируем уникальные значения и получаемые их словарные формы.
lemma=m.lemmatize(lemma) 

# определим самые частые леммы из столбца locality_name
print(Counter(lemma)) 

#список с леммами для удаления
outsiders = ['платформа', 'при', 'железнодорожной','садовое', 'садоводческое', 'некоммерческое','пансионат','поселок','посёлок','деревня','городского','городское','типа','село','станция','коттеджный','товарищество','станции','городской']

#разделяем строки locality_name_short по словам
dr['locality_name_short'] = dr['locality_name'].str.split(' ')
dr['locality_name_short']= dr['locality_name_short'].replace("ё","е")
# проверка разделения значений столбца 
#display(dr['locality_name_short']) 

#создаем собственную функцию для дальнейшей обработки столбца locality_name_short
def nef(df): 
    p=""
    for i in df:
        if i not in outsiders:
            p+=(" "+i)
    return p 
 
#применяем фунцию nef к столбцу        
dr['locality_name_short'] = dr['locality_name_short'].apply(nef) 
#достаточно малое количество строк с отсуствующим наименованием (0,2 %), можно удалить их.
dr.dropna(subset=['locality_name'],inplace=True)
#print(dr['locality_name'].isna().sum()) # проверка
 

#заполняем по группам по локации столбец расстояния до ближайщего парка.
for locality in dr['locality_name_short'].unique():
    median = dr.loc[dr['locality_name_short'] == locality, 'parks_nearest'].median()
    print(locality,median)
    dr.loc[(dr['parks_nearest'].isna()) & (dr['locality_name_short']==locality), 'parks_nearest']= median
    dr.loc[(dr['parks_nearest'].isna()), 'parks_nearest']= dr['parks_nearest'].median() 
print('Пропуски после:', dr['parks_nearest'].isna().sum())

#заполняем по группам по локации столбец расстояния до ближайщего водоема.
for locality in dr['locality_name_short'].unique():
    median = dr.loc[dr['locality_name_short'] == locality, 'ponds_nearest'].median()
    print(locality,median)
    dr.loc[(dr['ponds_nearest'].isna()) & (dr['locality_name_short']==locality), 'ponds_nearest']= median
    dr.loc[(dr['ponds_nearest'].isna()), 'ponds_nearest']= dr['ponds_nearest'].median()   
print('Пропуски после:', dr['ponds_nearest'].isna().sum())

#display(dr['parks_around3000'].isna().sum()) #проверка
#display(dr['ponds_around3000'].isna().sum()) #проверка

#расчитываем цену среднего метра, округляем то рубля, так как копейки в расчетах роли не сыграют.
dr['meter_price'] = round(dr['last_price']/dr['total_area'])

#проверка высокоэтажных домов
display(dr[dr['floors_total']>35]) 
#навряд-ли в Кронштадте есть небоскреб, так что заменяем значение на более логичное, в СПБ также самый высокий дом 38 этажей, а не 52 
dr['floors_total'] = dr['floors_total'].replace(60,6)
median_floors = dr.loc[dr['floors_total']!=52,'floors_total'].median()
dr['floors_total'] = dr['floors_total'].replace(52,median_floors)
#отсеим пустые значения dr['floors_total'], так как их малое количество не существенно для проводимого исследования. 
dr =dr.dropna(subset=['floors_total'])
# подсчитываем отношения площадей в отдельных столбцах.
dr['kitchen_cor'] = round(dr['kitchen_area']/dr['total_area'],2)
dr['living_cor'] = round(dr['living_area']/dr['total_area'],2)
display(dr['locality_name_short']) #проверка
dr.head()        #проверка

#для заполнения столбца по количеству дней жизни объявления задаем новый столбец, который определяет пропущена дата или нет.
dr['days_exposition_missing'] = dr['days_exposition'].isna().astype(int)
#суммируем и строим график
dr.groupby('first_day_exposition').agg({'days_exposition_missing':'sum'}).plot()
plt.xlabel('День размещения объявления'), plt.ylabel('Количество объявлений')
plt.title('Распределение объявлений\nбез days_exposition')
plt.show()

# пропущенные значения увеличивались к дате последних объявлений, что говорит о том, что на дату закрытия отчета эти квартиры ещё не были проданы
# заполним разницой последнего опубликованного объявления и датой публикации пропущенных по сроку объявлений Series dr['days_exposition'].
#подсчет пропусков
print('Пропуски до {}'.format(dr['days_exposition'].isna().sum()))
max_dt = dr['first_day_exposition'].max()
#заполняе пропуски разностью first_day_exposition и max_dt в днях
dr.loc[dr['days_exposition'].isna(), 'days_exposition'] = (max_dt - dr.loc[dr['days_exposition'].isna(), 'first_day_exposition']).dt.days
print('Пропуски после {}'.format(dr['days_exposition'].isna().sum()))

#заполняем по группам по локации столбец расстояния до центра города.
for locality in dr['locality_name_short'].unique():
    median = dr.loc[dr['locality_name_short'] == locality, 'cityCenters_nearest'].median()
    print(locality,median)
    dr.loc[(dr['cityCenters_nearest'].isna()) & (dr['locality_name_short']==locality), 'cityCenters_nearest']= median
    dr.loc[(dr['cityCenters_nearest'].isna()), 'cityCenters_nearest']= dr['cityCenters_nearest'].median() 
print('Пропуски после:', dr['cityCenters_nearest'].isna().sum())

#заполняем по группам по локации столбец расстояния до Пулково.
for locality in dr['locality_name_short'].unique():
    median = dr.loc[dr['locality_name_short'] == locality, 'airports_nearest'].median()
    print(locality,median)
    dr.loc[(dr['airports_nearest'].isna()) & (dr['locality_name_short']==locality), 'airports_nearest']= median
    dr.loc[(dr['airports_nearest'].isna()), 'airports_nearest']= dr['airports_nearest'].median() 
print('Пропуски после:', dr['airports_nearest'].isna().sum())
```

### Оценка стоимости квартир

```python
dr['last_price'].plot.box(grid=True)
plt.ylim(1,14000000)
plt.show()

dr['last_price'].hist(bins=10, figsize=(20,5))
plt.xlabel('Цена за квартиру'), plt.ylabel('Количество объявлений')
plt.show()

dr['last_price'].hist(bins=10, range=(2000000,16000000),figsize=(20,5))
plt.xlabel('Цена за квартиру'), plt.ylabel('Количество объявлений')
plt.show()

#расчет IQR
q1_price = dr['last_price'].quantile(0.25)
q3_price = dr['last_price'].quantile(0.75)
iqr_price = q3_price - q1_price 
low_board_price = q1_price - (1.5 * iqr_price)
high_board_price = q3_price + (1.5 * iqr_price)
print(f'Выбросы начинаются с {high_board_price} рублей.')

max_board_price = dr['last_price'].quantile(0.99)
min_board_price = 1-(dr['last_price'].quantile(0.99))
#боьше всего предложений на квартиры в районе 4000000 рублей.
print(f'Самые часто встречающиеся квартиры ценой между {q1_price} и {q3_price} рублей.')
#граница размаха слишком много откидывает объявлений, поэтому ограничемся отсеиванием 1 % объявлений, для более широкого исследования.
dr = dr[(dr['last_price']<=max_board_price)&(dr['last_price']>=min_board_price)]

#display(dr.tail()) #проверка
#display(dr.info()) #проверка

#проверка идеи, что 4 млн. средняя цена для однокомнатной квартиры 
dr.groupby('rooms').agg({'last_price':'mean'})
```

#### Вывод

<dl>На момент исследования самые частые и видимо ходовые предложения были на 4 млн. рублей, квартиры ценой до 7 млн также были востребованы. В основном цена в 3.5.-4 млн. рублей установлена на 1-комнтаные квартиры, которые и являются самым частым предложением на рынке.</dl>

### Оценка площади квартир

```python
dr['total_area'].plot.box(grid=True)
plt.ylim(1,160)
plt.show()
#выбросы определены

dr['total_area'].hist(bins=450, range =(0,400), figsize=(20,5))
plt.xlabel('Общая площадь'), plt.ylabel('Количество объявлений')
plt.show()

#расчет IQR
q1_area = dr['total_area'].quantile(0.25)
q3_area = dr['total_area'].quantile(0.75)
iqr_area = q3_area - q1_area 
low_board_area = q1_area - (1.5 * iqr_area)
high_board_area = q3_area + (1.5 * iqr_area)
print(f'Выбросы начинаются с {high_board_area} кв. метров.' )
display(dr['total_area'].describe())#проверка
#чаще всего встречаются 50 метровые квартиры.
print(f'Самые частые метражи квартир между {q1_area} и {q3_area} кв. метров.')

max_board_area = dr['total_area'].quantile(0.99)
min_board_area = 1-(dr['total_area'].quantile(0.99))
dr = dr[(dr['total_area']<=max_board_area)&(dr['total_area']>=min_board_area)]

#display(dr.tail()) #проверка
#display(dr.info()) #проверка

#проверка идеи, что 40-50 м. средняя площадь для однокомнатной квартиры 
dr.groupby('rooms').agg({'total_area':'mean'})
```

#### Вывод
