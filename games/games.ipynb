### Предобработка данных

```python
# переименование названия столбцов для упрощения обращения к ним
df = df.rename(columns={'Year_of_Release': 'year_of_release','NA_sales': 'na_sales','EU_sales': 'eu_sales','JP_sales': 'jp_sales', 'User_Score': 'user_score', 'Other_sales':'other_sales', 'Critic_Score':'critic_score','Name':'name','Platform':'platform','Genre':'genre','Rating':'rating'})

# замена пропущенных значений на 'unknown'
df['rating'] = df['rating'].fillna('unknown')

# список столбцов для приведения к нижнему регистру
list_one=['name', 'platform', 'genre', 'rating']
# цикл для приведения к нижнему регистру столбцов из списка list_one
for lower in df: 
    if lower in list_one:
        df[lower] = df[lower].str.lower()
        
# проверка         
display(df['name'].value_counts().sort_values()) 

# проверка теории, что количество пропусков в столбцах name и genre взаимосвязано.
print('Количество пропусков в столбцах name и genre:',len(df[df['name'].isna() & df['genre'].isna()]))

# так как количество пропусков в столбцах genre, name, year_of_release малое, можно почистить от них дф.
df = df.dropna(subset = ['name','genre','year_of_release'])

# проверка убранных успешно значений в дф.
print('Количество пропусков в столбцах name и genre:',len(df[df['name'].isna() & df['genre'].isna()]))

# оценка полных дубликатов, таковые отсутствуют.
print('Количество полных дубликатов:',df.duplicated().sum())

# так как одна игра может выходить на нескольких платформах одновременно,
# искать дубликаты по всему дф по названию игры не целесобразно, поэтому мы проверим повторы по платформам и названию игр.
print('Количество повторов по игре и платформе:', df[['name', 'platform']].duplicated().sum())

# оценим найденные повторы зрительно.
display(df[df.duplicated(subset=['name','platform'], keep=False)])

# нужно убить строку дубликат 16230, остальные два повтора – не дубликаты, у них разные годы выпуска.
df = df.drop(16230)

# проверка
#print(df[df.duplicated(subset=['name','platform'], keep=False)])
```

```python
# преобразование типов данных.
df.loc[df['user_score'] == 'tbd', 'user_score'] = float('nan')
df['user_score'] = df['user_score'].astype(float)

df['year_of_release'] = pd.to_datetime(df['year_of_release'],format ='%Y')

# добавление столбца total_sales с суммарными продажами во всех регионах.
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']

# проверка
display(df.info())
```

#### Вывод
<dl>Для нашего удобства при работе с данными мы провели пару операций над массивом:</dl>
<li>Был создан общий столбец продаж, который поможет оценить полную картину по регионам.</li>
<li> Приведены названия игр, платформ, жанров, возрастного рейтинга к нижнему регистру и таким образом исключены дубликаты.</li>
<li> Также обработаны названия столбцов.</li>
<li> Типы данных стали соответствовать своим значениям.</li>
    
Оценив предварительно исследуемые данные мы обнаружили незначительное количество пропусков в `столбцах genre`, name и `year_of_release`, от которых мы поспешно почиститли таблицу. К сожалению, добыть пропущенную информацию мы не можем, а пропуски только исказят исследование, так что меры применяются категорические. Хотя будь такая возможность нам бы стоило уточнить у дата-инженера не потерялись ли данные при выгрузке, ну или проверить самих себя.<dl></dl>
    Пока что пустые значения в `critic_score` и `user_score` мы сохраним, так как они затрагивают слишком большое количество строк для предварительной обработки и их удаления, но в дальнейшем при проведении t-test нам придется с ними попращаться.<dl></dl>
    

### Характиеристика общего периода исследования 


```python
# гистограмма по общему количеству игр в год
df['year_of_release'].hist(bins=37,figsize=(20,5), color='maroon', histtype='step')
plt.xlabel('Год'), plt.ylabel('Количество выпущенных игр')
plt.suptitle('Гистограмма количества выпущенных игр по годам')
plt.show()

# создадим круговую диаграмму для выделения наиболее популярных платформ своего времени.
df.groupby('platform').agg({'total_sales':'sum'}).sort_values(by='total_sales',ascending=False).head(10).plot(kind='pie', y = 'total_sales',figsize=(10,10), autopct='%1.1f%%')
plt.title('Круговая диаграмма продаж игр по платформам')
plt.show()
# 6 лидеров в продажах по всем годам ps2, x360, ps3, wii, ds, ps.

df.query('platform in ["ps2", "x360","wii", "ps", "ps3", "ds","ps4","gba","psp","gb"]').pivot_table(index='year_of_release', columns = 'platform', values='total_sales', aggfunc='sum').plot(grid=True, figsize=(20, 5))
plt.xlabel('Год'), plt.ylabel('Сумма продаж за год')
plt.suptitle('Гистограмма продаж выпущенных игр по годам на топовых платформах')
plt.show()

# гистрограммы оценки связи отызывов и продаж игр.
df.plot(kind='scatter',x='critic_score' ,y='total_sales', grid=True, alpha=0.3)
plt.xlabel('Оценка критиков'), plt.ylabel('Количество проданных игр (млн.)')
plt.suptitle('Диаграмма рассеяния зависимости количества проданных игр от оценки критиков')
plt.show()

df.plot(kind='scatter',x='user_score' ,y='total_sales', grid=True, alpha=0.3)
plt.xlabel('Оценка игроков'), plt.ylabel('Количество проданных игр (млн.)')
plt.suptitle('Диаграмма рассеяния зависимости количества проданных игр от оценки игроков')
plt.show()

```

### Определение актуального периода для прогнозирования

```python
# формирование сводной таблицы по году выпуска игры.
pd.set_option('display.max_columns',40)
df.pivot_table(index='platform', columns='year_of_release', values='total_sales',aggfunc='sum')


```

```python
# расчет актуального периода, т.е. половины от среднего времени жизни платформы, пока спрос на нее поднимается.
# минимальные значимые продажи в миллионах
min_sales = 10
#суммарные продажи в год выше порога
report = df.groupby(['platform', 'year_of_release']).agg({'total_sales':'sum'}).query('total_sales > @min_sales')
# среднее число лет выше порога
report.reset_index().groupby('platform').agg({'year_of_release':'nunique'}).mean()/2
```

### Определение самой популярной платформы и жанра

```python
# всего продаж за актуальный период по всем платформам.
print('Всего продаж за актуальный период по всем платформам:',df[df['year_of_release'].isin(['2014','2015','2016'])]['total_sales'].sum())
# оценим актулаьные на данный момент платформы по продажам игр.
df_actually = df[df['year_of_release'].isin(['2014','2015','2016'])]
# display(df_actually) # проверка
df_actually.pivot_table(index='year_of_release', columns = 'platform', values='total_sales', aggfunc='sum').plot(grid=True, figsize=(20, 5))
plt.xlabel('Год'), plt.ylabel('Сумма продаж за год')
plt.suptitle('Гистограмма продаж выпущенных игр по годам за актуальный период')
plt.show()
# на даннй момент лидирует ps4, xone, 3ds

# формирование сводной таблицы продаж по году выпуска игры.
pd.set_option('display.max_columns',40)
display(df_actually.pivot_table(index='platform', columns='year_of_release', values='total_sales',aggfunc='sum'))

# продажи за 3 года.(Кусок от общих продаж)
df_actually['year_of_release'].hist(bins=3,figsize=(7,3), color='steelblue', histtype='step')
plt.xlabel('Год'), plt.ylabel('Количество выпущенных игр')
plt.suptitle('Гистограмма количества выпущенных игр по годам')
plt.show()

# зависимости продаж от оценок игроков и критиков. 
df_actually.plot(kind='scatter',x='critic_score' ,y='total_sales', grid=True, alpha=0.3)
plt.xlabel('Оценка критиков'), plt.ylabel('Количество проданных игр (млн.)')
plt.suptitle('Диаграмма рассеяния зависимости количества проданных игр от оценки критиков')
plt.show()

df_actually.plot(kind='scatter',x='user_score' ,y='total_sales', grid=True, alpha=0.3)
plt.xlabel('Оценка игроков'), plt.ylabel('Количество проданных игр (млн.)')
plt.suptitle('Диаграмма рассеяния зависимости количества проданных игр от оценки игроков')
plt.show()


    
```

### Проверка гипотез

```python
display(df_actually.info())

#удаляем пустые значения перед исследованием
df_actually = df_actually.dropna(subset=['critic_score','user_score'])

# гипотеза "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые":
#H0: средние пользовательские рейтинги платформ Xbox One и PC равны
#H1: средние пользовательские рейтинги платформ Xbox One и PC не равны
#По результатам стат-теста одна из этих гипотез подтвердится.
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

sample_1 = df_actually.loc[df_actually['platform']=='xone','user_score']
sample_2 = df_actually.loc[df_actually['platform']=='pc','user_score']

results = st.ttest_ind(
    sample_1, 
    sample_2)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу. Cредние пользовательские рейтинги платформ Xbox One и PC не равны.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Средние пользовательские рейтинги платформ Xbox One и PC равны.")
    
#среднние пользовательские оценки по платформам       
print('Средние пользовательские рейтинги платформы Xbox One: ', (round(df_actually.loc[df_actually['platform']=='xone','user_score'].mean(),2)))
print('Средние пользовательские рейтинги платформы PC: ', (round(df_actually.loc[df_actually['platform']=='pc','user_score'].mean(),2)))
    
# гипотеза "Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.":
#H0: средние пользовательские рейтинги жанров Action и Sports равны
#H1: средние пользовательские рейтинги жанров Action и Sports не равны
#По результатам стат-теста одна из этих гипотез подтвердится.
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

sample_1 = df_actually.loc[df_actually['genre']=='action','user_score']
sample_2 = df_actually.loc[df_actually['genre']=='sports','user_score']

results = st.ttest_ind(
    sample_1, 
    sample_2)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу. Cредние пользовательские рейтинги жанров Action и Sports не равны.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Средние пользовательские рейтинги жанров Action и Sports равны.")
    
#среднняя пользовательские оценки по жанрам    
print('Cредние пользовательские рейтинги жанра action: ',  (round(df_actually.loc[df_actually['genre']=='action','user_score'].mean(),2)))
print('Cредние пользовательские рейтинги жанра sports: ',  (round(df_actually.loc[df_actually['genre']=='sports','user_score'].mean(),2)))

        
```

#### Вывод
<dl></dl>При проведении стат-теста нам удалось выявить статистически значимые различия между жанрами Action и Sports. Напротив, статистически значимые различия между средними пользовательскими рейтингами платформ Xbox One и PC выявить не удалось.
